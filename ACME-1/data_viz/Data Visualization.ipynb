```python
def problemOne():
    '''
    Order and plot the data as a horizontal bar chart.
    '''
    male = np.array([179.2,160.0,177.8,178.9,178,176,172.5,165.6,170.8,183.2,182.4,164,163.6,175.4,174,176.1,165.7])
    female = np.array([167.6,142.2,164.5,165.3,165,164,158,154.9,157.4,168.4,168,151,151.4,164,158.9,162.1,155.2])
    country = np.array(['Austria', 'Bolivia', 'England', 'Finland', 'Germany', 'Hungary', 'Japan','North Korea', 'South Korea', 'Montenegro', 'Norway', 'Peru','Sri Lanka', 'Switzerland', 'Turkey', 'U.S.', 'Vietnam'])
    sort_index = np.argsort(male)
    country = country[sort_index]
    male = male[sort_index]
    pos = np.arange(17)+.5
    
    plt.subplot(121)
    plt.title("Men")
    plt.barh(pos, male, align='center')
    plt.yticks(pos, country)
    
    plt.subplot(122)
    plt.title("Women")
    ort_index = np.argsort(female)
    country = country[sort_index]
    female = female[sort_index]
    plt.barh(pos, female, align='center')
    plt.yticks(pos, country)
    plt.tight_layout()
    plt.show()

problemOne()
```

```python
def problemTwo():
    '''
    Plot some histograms with white reference lines.  Do this for 20 bins,
    10 bins, 5 bins, and 3 bins
    '''
    normal_numbers = normal(size=1000)
    plt.subplot(141)
    plt.grid(True, color='w', linestyle='-')
    plt.title("20 bins")
    plt.hist(normal_numbers, 20)
    plt.subplot(142)
    plt.title("10 bins")
    plt.grid(True, color='w', linestyle='-')
    plt.hist(normal_numbers, 10)
    plt.subplot(143)
    plt.title("5 bins")
    plt.grid(True, color='w', linestyle='-')
    plt.hist(normal_numbers, 5)
    plt.subplot(144)
    plt.title("3 bins")
    plt.grid(True, color='w', linestyle='-')
    plt.hist(normal_numbers, 3)
    plt.title("histogram")
    plt.tight_layout()
    plt.show()

problemTwo()
```

```python
def problemFive():
    '''
    Plot a contour map of z = sin(x) + sin(y) where x is in [0,12*pi] and
    y is in [0,12*pi]
    '''
    n = 400
    xran = np.linspace(0,12*np.pi,n)
    yran = np.linspace(0,12*np.pi,n)
    X, Y = np.meshgrid(xran, yran)
    F = np.sin(X)+np.sin(Y)
    plt.contourf(X, Y, F,cmap=plt.get_cmap('afmhot'))
    
problemFive()
    
```

```python
def problemSeven():
    '''
    Change the surface to a heatmap or a contour plot.  Return a string
    of the benefits of each type of visualization.
    '''
    x = np.linspace(-2*np.pi,2*np.pi,num=400)
    y = np.linspace(-2*np.pi,2*np.pi,num=400)
    X, Y = np.meshgrid(x,y)
    Z = np.exp(np.cos(np.sqrt(X**2 + Y**2)))
    plt.subplot(121)
    plt.contourf(X, Y, Z, cmap=plt.get_cmap('afmhot'))
    plt.subplot(122)
    plt.contour(X, Y, Z, cmap=plt.get_cmap('afmhot'))
    plt.tight_layout()
    plt.show()
    
problemSeven()
```

```python
def problemEight():
    '''
    Plot y = x^2 * sin(x) where x is in [0,100] and adjust to y limit to be
    [-10^k,10^k] for k = 0,1,2,3,4.
    '''
    x = np.linspace(0,101,1000)
    y = x**2 * np.sin(x)
    
    plt.subplot(321)
    plt.ylim(-10**0, 10**0)
    plt.plot(x, y,lw=2)
    
    plt.subplot(322)
    plt.ylim(-10**1, 10**1)
    plt.plot(x, y,lw=2)
    
    plt.subplot(323)
    plt.ylim(-10**2, 10**2)
    plt.plot(x, y,lw=2)
    
    plt.subplot(324)
    plt.ylim(-10**3, 10**3)
    plt.plot(x, y,lw=2)
    
    plt.subplot(325)
    plt.ylim(-10**4, 10**4)
    plt.plot(x, y,lw=2)
    
    plt.tight_layout()
    plt.show()

problemEight()
```

```python
def problemNine():
    '''
    Simplify one of your previous graphs.
    '''
    n = 400
    xran = np.linspace(0,12*np.pi,n)
    yran = np.linspace(0,12*np.pi,n)
    X, Y = np.meshgrid(xran, yran)
    F = np.sin(X)+np.sin(Y)
    spacing = [-2, -1.5, -1, -.5, .001, .5, 1, 1.5, 2]
    plt.contourf(X ,Y ,F ,spacing ,cmap=plt.get_cmap('cool'))
    plt.xlim(0,15)
    plt.ylim(0,15)
    plt.title(r'Graph of sin(x)+sin(y)')
    plt.grid()
    plt.show()

problemNine()
```

```python
def problemTen():
    '''
    Plot the Bernstein polynomials for v,n in [0,3] as small multiples
    and as the cluttered version.
    '''
    
    burns = lambda v,n,x: comb(n, v)*x**v*(1-x)**(n-v)
    
    n = 4
    x = np.linspace(0,1,600)
    for i in xrange(n+1):
        for j in xrange(i+1):
            plt.plot(x, burns(j,i,x))
    plt.show()
    
problemTen()
```
