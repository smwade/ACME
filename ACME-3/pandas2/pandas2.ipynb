## Problem 1

### Nottem
The average air temperatures at Nottingham Castle in Fahrenheit for 20 years.

```python
nottem = data("nottem")
data("nottem", show_doc=True)
```

```python
nottem.head()
```

```python
nottem.plot(x='time', y='nottem')
plt.title("Temperature in Nottingham, 1920-1939")
plt.ylabel("Temp (F.)")
plt.xlabel("Time (Years)")
plt.show()
```

*Source: Anderson, O. D. (1976) _Time Series Analysis and Forecasting: The Box-Jenkins
approach._ Butterworths. Series R*

This visualization shows a time series of the variation of temperature. It is appropriate to use a line graph to show the periodic nature of the weather.

### VADeaths
Death rates per 1000 in Virginia in 1940.

```python
VADeaths = data("VADeaths")
data("VADeaths", show_doc=True)
```

```python
VADeaths.head()
```

```python
VADeaths[['Urban Male','Urban Female']].plot(kind='bar')
plt.title("Urban Death Rates in Verginia, 1940")
plt.xlabel("Age (years)")
plt.ylabel("Rate")
VADeaths[['Rural Male','Rural Female']].plot(kind='bar')
plt.title("Rural Death Rates in Verginia, 1940")
plt.xlabel("Age (years)")
plt.ylabel("Rate")

plt.show()
```

*Source: Molyneaux, L., Gilliam, S. K., and Florant, L. C.(1947) Differences in
Virginia death rates by color, sex, age, and rural or urban residence.
_American Sociological Review_, **12**, 525-535\.*

This visualization shows the relation between the deaths of males and females. The two graphs are comparing in rural and urban environments.

### Arbuthnot
 Ratios of male to female births in London from 1629-1710.

```python
Arbuthnot = data("Arbuthnot")
data("Arbuthnot", show_doc=True)
```

```python
Arbuthnot.head()
```

```python
Arbuthnot.plot(x='Year',y=['Males', 'Females'])
plt.title('Births in London, 1629-1710')
plt.ylabel('Number of People')
Arbuthnot.plot(x='Year', y='Mortality')
plt.title('Mortality in London, 1629-1710')
plt.ylabel('Number of People')
plt.show()
```

*Source: Arbuthnot, John (1710). "An argument for Devine Providence, taken from the
constant Regularity observ'd in the Births of both Sexes," _Philosophical
transactions_, 27, 186-190. Published in 1711.*

The first visualization shows the relation between male and female births.  The second shows the mortality over time.

### Road
Road accident deaths in the United States.

```python
road = data("road")
data("road", show_doc=True)
```

```python
road.head()
```

```python
road['deaths'].plot(kind='box')
plt.title('Deaths in Road Accidents, United States')
plt.ylim(0,2500)
plt.ylabel("Number of People")
plt.show()
```

*Source: Imperial College, London M.Sc. exercise*

This box chart for road deaths in the US shows the high variance for deaths in the upper 2 quartiles.
