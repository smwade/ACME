```python
import pandas as pd
from pydataset import data
from matplotlib import pyplot as plt
import seaborn
plt.style.use('fivethirtyeight')
```

```python
diamonds = data('diamonds')
```

```python
diamonds.head()
```

```python
cut = diamonds.groupby('cut')
cut.describe()
```

```python
cut = diamonds.groupby('cut')
for x in diamonds['cut'].unique():
    cut.get_group(x)['color'].value_counts().plot(kind='bar')
    plt.xlabel("Color")
    plt.ylabel("Amount")
    plt.title("Amount of Colors in {}".format(x))
    plt.show()   
```

*Source: R dataset Diamonds*

**Diamond Analysis**:  THe above table shows the different statistics for each group. The average cost and variance are interesting.  The graphs are showing. The different types of colors with respect to quality.  These show that there is not much correlation between color and quality. G is always the most common color and J is always the least common color.

## World Phone Data, 1951-1961

```python
phones = data('WorldPhones')
data('WorldPhones', show_doc=True)
```

```python
phones.head()
```

The data

```python
phones.describe()
```

```python
phones.plot()
plt.title('Number of World Phones Over Time, 1951-1961')
plt.xlabel('Year')
plt.ylabel('Number of Phones')
plt.show()
```

*Source: AT&T (1961) _The World's Telephones_.*

This models the growth of phones in these areas over time.  This clearly shows how dominate the rates of growth in N.America and Europe are.  While on the lower side of the spectrum there is growth, but it seems constant in comparison.

```python
phones.mean().plot(kind='bar')
plt.title("Mean Number of Phones, 1951-1961")
plt.xlabel("Region")
plt.ylabel("Number of Phones")
plt.show()
```

*Source: AT&T (1961) _The World's Telephones_.*

This graph illustrates the mean number of phones over the years 1951-1962. These graphs correspond to the more technically developed countries in the world.

```python
phones[['Europe', 'N.Amer']].plot(kind='box')
plt.title("America vs. Europe in Number of Phones")
plt.ylabel('Phones')
plt.show()
```

*Source: AT&T (1961) _The World's Telephones_.*

There is much more variance in the number of phones in N. America than Europe.  This could be used to explain the rapid growth in N.America over this time.

```python
phones.pct_change()
```

This table illustrates the percent change in number of phones over time.
